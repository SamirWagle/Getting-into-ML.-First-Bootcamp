```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
data_path = "https://www.statlearning.com/s/Advertising.csv" 

# Read the CSV data from the link
data_df = pd.read_csv(data_path,index_col=0)

# Print out first 5 samples from the DataFrame
data_df
```

```python
fig , axes = plt.subplots(1,3)
fig.set_figheight(4)
fig.set_figwidth(15)
axes[0].scatter(data_df["TV"], data_df["sales"], color="red", marker=".")
axes[0].set_xlabel("TV")
axes[0].set_ylabel("sales")

axes[1].scatter(data_df["radio"], data_df["sales"], color="blue", marker=".")
axes[1].set_xlabel("radio")
axes[1].set_ylabel("sales")

axes[2].scatter(data_df["newspaper"], data_df["sales"], color="purple", marker=".")
axes[2].set_xlabel("newspaper")
axes[2].set_ylabel("sales")

```

```python
fig , axes = plt.subplots(1,3)
fig.set_figheight(4)
fig.set_figwidth(15)

# function for training model and plotting
def train_plot(data_df, feature, ax, col):
    X =data_df[[feature]].values
    Y = data_df[['sales']].values
    sum_Y = np.sum(Y)
    sum_X = np.sum(X)
    sum_XY =np.sum(X*Y)
    sum_X_squared = np.sum(X**2)
    n= len(X)
    
    a =np.array([[sum_X,n],[sum_X_squared,sum_X]])
    b= np.array([sum_Y,sum_XY])
    m = np.linalg.solve(a,b)[0]
    c = np.linalg.solve(a,b)[1]
    y_hat = np.dot(X,m)+c
    print(m)
    
    reg = linear_model.LinearRegression()
    reg.fit(X,Y)
    print(reg.coef_,reg.intercept_,c)
    y_hat = np.dot(X,reg.coef_)+reg.intercept_
    ##Plot the regression line
    ax.scatter(data_df[feature], data_df["sales"], color=col, marker=".")
    ax.plot(X, y_hat, color="black")
    ax.set_xlabel(feature)
    ax.set_ylabel("sales")
    ax.set_title(("$y$ = %3f + %3f$x$" %(c, m)))


    
# Train model using TV data to predict sales
train_plot(data_df, "TV", axes[0], "red")

# Train model using radio data to predict sales
train_plot(data_df, "radio", axes[1], "green")

# Train model using newspaper data to predict sales
train_plot(data_df, "newspaper", axes[2], "blue")

plt.show()
    
```
