```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python

x = np.linspace(0,5,11)
x
```

```python
y = x**2
y
```

```python
plt.xlabel('x')
plt.ylabel('y')
plt.title(r'$y=x^2$')
plt.plot(x,y)
plt.show()
```

```python
x
```

```python
plt.title('Graphs')
plt.subplot(1,2,2)
plt.plot(x,y,'r*:')
plt.xlabel('x')
plt.ylabel(r'$x^2$')
plt.subplot(1,2,1)
plt.xlabel('y')
plt.ylabel('x')

plt.plot(y,x,'g*:')
plt.tight_layout()
plt.show()
```
